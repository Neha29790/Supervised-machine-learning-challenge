## Split the Data into Training and Testing Sets

```python
# Split the data into X_train, X_test, y_train, y_test
# DEFINE FEATURES (X) AND TARGET (Y) SETS
X = lending_df.drop("loan_status", axis = 1)
y = lending_df["loan_status"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Create, Fit and Compare Models

Create a Logistic Regression model, fit it to the data, and print the model's score. Do the same for a Random Forest Classifier. You may choose any starting hyperparameters you like. 

Which model performed better? How does that compare to your prediction? Write down your results and thoughts in the designated markdown cell.

```python
# Train a Logistic Regression model and print the model score
logreg_model = LogisticRegression(max_iter=10000)
logreg_model.fit(X_train, y_train)
print(f"Training Data Score: {logreg_model.score(X_train, y_train)}")
print(f"Testing Data Score: {logreg_model.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
ranfor_model = RandomForestClassifier(random_state = 42, n_estimators = 10000)
ranfor_model.fit(X_train_scaled, y_train)

```

```python
print(f'Training Score: {ranfor_model.score(X_train_scaled, y_train)}')
print(f'Testing Score: {ranfor_model.score(X_test_scaled, y_test)}')

```

 The Logistic Regression model performed more accurately on the test data set than the Random Forest model and it took 0.8s to run altogether.The Random Forest model performed slightly more accurately on the training set than the Logistic Regression model and took 7m 40.6s to run.
